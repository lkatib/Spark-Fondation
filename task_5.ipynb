```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the dataset
data = pd.read_csv('matches.csv')

# Visualize the distribution of a numerical variable using a histogram
plt.figure(figsize=(8, 6))
sns.histplot(data['win_by_runs'], kde=True)
plt.title('Distribution of Win by Runs')
plt.xlabel('Win by Runs')
plt.ylabel('Frequency')
plt.show()

# Visualize the relationship between two numerical variables using a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='win_by_runs', y='win_by_wickets')
plt.title('Relationship between Win by Runs and Win by Wickets')
plt.xlabel('Win by Runs')
plt.ylabel('Win by Wickets')
plt.show()

# Visualize the relationship between a categorical variable and a numerical variable using a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='toss_decision', y='win_by_runs')
plt.title('Relationship between Toss Decision and Win by Runs')
plt.xlabel('Toss Decision')
plt.ylabel('Win by Runs')
plt.show()

# Visualize the distribution of a categorical variable using a bar plot
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='toss_winner')
plt.title('Distribution of Toss Winners')
plt.xlabel('Toss Winners')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()

# Visualize a time series variable using a line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='date', y='win_by_runs')
plt.title('Time Series Analysis of Win by Runs')
plt.xlabel('Date')
plt.ylabel('Win by Runs')
plt.xticks(rotation=90)
plt.show()
```

```python
team_matches = data2['team1'].value_counts() + data2['team2'].value_counts()
```

```python
team_wins = data2['winner'].value_counts()
```

```python
team_losses = team_matches - team_wins
```

```python
team_win_percentages = (team_wins / team_matches) * 100
```

```python
most_successful_teams = team_win_percentages.sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=most_successful_teams.index, y=most_successful_teams.values)
plt.title('Winning Percentages of Teams')
plt.xlabel('Teams')
plt.ylabel('Winning Percentage')
plt.xticks(rotation=45)
plt.show()
```

```python
batting_averages = data1.groupby('batsman')['batsman_runs'].sum() / data1.groupby('batsman')['player_dismissed'].count()
bowling_averages = data1[data1['dismissal_kind'].notna()].groupby('bowler')['total_runs'].sum() / data1[data1['dismissal_kind'].notna()].groupby('bowler')['player_dismissed'].count()
total_runs = data1.groupby('batsman')['batsman_runs'].sum()
total_wickets = data1[data1['dismissal_kind'].notna()].groupby('bowler')['player_dismissed'].count()
```

```python
top_batting_averages = batting_averages.sort_values(ascending=False)
top_bowling_averages = bowling_averages.sort_values()
top_runs = total_runs.sort_values(ascending=False)
top_wickets = total_wickets.sort_values(ascending=False)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=top_runs.index[:10], y=top_runs.values[:10])
plt.title('Top 10 Players - Total Runs')
plt.xlabel('Players')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x=top_wickets.index[:10], y=top_wickets.values[:10])
plt.title('Top 10 Players - Total Wickets')
plt.xlabel('Players')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45)
plt.show()
```
